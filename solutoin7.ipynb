# CALCULATOR USING FUNCTIONS

```python
# Function for Addition 

def add(a, b):
    return a + b

# Function for Subtraction

def subtract(a, b):
    return a-b

# Function for Multiplication

def multiply(a, b): 
    return a*b

# Function for Division 

def divide(a, b):
    if b == 0:
        return "Error division by zero is not allowed"
    else:
        return a/b
    
# Function for Power

def power(a, b):
    return a ** b
        
# Function for Module 

def module(a, b):
    return a % b

# Operations in the Calculator 

def operations():
    print("Select Operation: ")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Power")
    print("6. Module")
    
# Taking input from the user for the Operation

choice = input("Enter your choice (1,2,3,4,5,6): ")
if choice in ["1", "2", "3", "4", "5", "6"]:
    num1 = float(input("Enter your first number: "))
    num2 = float(input("Enter your second number: "))
    
if choice == "1":
    print(f"The Addition of {num1} and {num2} is {add(num1, num2)}.")
    
elif choice == "2":
    print(f"The Subtraction of {num1} and {num2} is {subtract(num1, num2)}.")
    
elif choice == "3":
    print(f"The Multiplication of {num1} and {num2} is {multiply(num1, num2)}.")
    
elif choice == "4":
    print("The Division of {num1} and {num2} is {divide(num1, num2}.")
    
elif choice == "5":
    print(f"The Power of {num1} and {num2} is {power(num1, num2)}.")
    
elif choice == "6":
    print(f"The Module of {num1} and {num2} is {module(num1, num2)}.")

else: 
    print("Invalid input, Please enter a valid choice")

```
